```python
def note(x):
    if x == "Yes":
        return 1
    elif x == "No":
        return 0
    elif x =="In progress":
        return 0.5
    elif x =="Intending to undertake in future":
        return 0.25
    elif x == "Not intending to undertake":
        return 0
    elif x == "Do not know":
        return 0
    else:
        return 0
    
```

```python
df_clean['note'] = df_clean['Response Answer'].apply(note)
```
